```python
# Perfect sensor
def sense(x):
    return x
```

```python
def simulate(Δt, robot, u):
    robot += Δt * u
    return robot
```

```python
def control(t, pos):
    ### WRITE YOUR CONTROL POLICY HERE:
    
    ux = -sin(t)
    uy = cos(t)
    uz = .01
    return array([ux, uy, uz])
```

```python
tf = 10.
Δt = 0.01    # Time step
time = linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# Initial conditions
robot = array([2.0, 0.0, 0.0])
x_log = [copy(robot)]

for t in time:
    pos = sense(robot)
    u = control(t, pos)
    #u[1] += .5
    robot = simulate(Δt, robot, u)
    x_log.append(copy(robot))
    
x_log = array(x_log)
```

```python
ax = plt.axes(projection ='3d')
ax.plot3D(x_log[:,0], x_log[:,1], x_log[:,2])
```

```python
import matplotlib.pyplot as plt
from matplotlib import animation 
from IPython.display import HTML
    


fig, ax = plt.subplots()

ax = plt.axes(projection ='3d')

def animate(t):
    ax.clear()
    
    # Path
    ax.plot3D(x_log[:,0], x_log[:,1], 'r--')
    
    # Initial conditions
    ax.plot3D(x_log[t,0], x_log[t,1], 'bo')
    
    

anim = animation.FuncAnimation(fig, animate, frames=len(time), interval=60)

HTML(anim.to_jshtml())
```
